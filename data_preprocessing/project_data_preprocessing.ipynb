## 4. Категоризация данных
    
Произведена категоризация данных:
- Разделены все игры по оценкам пользователей и выдеены такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

```python
df_actual.sort_values(by=['name', 'platform']).head(50)
```

Ну блин, что делать с пропусками в рейтинге не понятно их слишком много, категоризировать игру на определенной платформе по рейтингу на другой не до конца правильно, искать какое то среднее на других площадках для этой игры тоже неверно, это с условием того что игра может не иметь рейтинга ни на одной из площадок, или быть выпущена тольокк на одной платформе и не иметь рейтинга.

При этом ну неверно убирать так много данных

Было принято решение выделить четвертую категориюне имеет оценки для того чтобы оставить как можно большее строк в датасете ведь игры были выпущениы и удаление строк повлияет на результат при получении топ-7 платформ, при этом оставить разделение на категории максимально объективным

```python
def user_rate_to_cat(row):
    if row['user_score'] <= 10 and row['user_score'] >= 8:
        return "высокая оценка"
    elif  row['user_score'] < 8 and row['user_score'] >= 3:
        return "средняя оценка"
    elif  row['user_score'] < 3 and row['user_score'] >= 0:
        return "низкая оценка"
    else: return "нет оценки"
df_actual['category_user'] = df_actual.apply(user_rate_to_cat, axis=1)
df_actual.info()
```

```python
def critic_rate_to_cat(row):
    if row['critic_score'] <= 100 and row['critic_score'] >= 80:
        return "высокая оценка"
    elif  row['critic_score'] < 80 and row['critic_score'] >= 30:
        return "средняя оценка"
    elif  row['critic_score'] < 30 and row['critic_score'] >= 0:
        return "низкая оценка"
    else: return "нет оценки"
df_actual['category_critic'] = df_actual.apply(critic_rate_to_cat, axis=1)
df_actual.info()
```

Категории по ррейтингу пользователей

```python
group_user = df_actual.groupby('category_user')['name'].agg('count').sort_values(ascending=False)
print(group_user) 
```

Категории по рейтингу критиков

```python
group_critic = df_actual.groupby('category_critic')['name'].agg('count').sort_values(ascending=False)
print(group_critic) 
```

топ-7 платформ по количеству игр, выпущенных за весь актуальный период.

```python
top_7 = df.groupby('platform')['name'].agg('count').sort_values(ascending=False).head(7)
print(top_7) 
```
